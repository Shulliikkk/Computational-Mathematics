```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def aver(data):
  return np.sum(data) / np.size(data)
def MNK(x, y):
  k = (aver(x * y) - aver(x) * aver(y)) / (aver(x**2) - aver(x)**2)
  b = aver(y) - k * aver(x)
  return k, b

df_4 = pd.read_csv('BDF_order_4.csv')
df_5 = pd.read_csv('BDF_order_5.csv')

err_4 = df_4['err'].to_numpy()
h_4 = df_4['h'].to_numpy()
err_5 = df_5['err'].to_numpy()
h_5 = df_5['h'].to_numpy()

k_4, b_4 = MNK(np.log(h_4[:4]), np.log(err_4[:4]))
print('Slope coefficient 4:', k_4)
k_5, b_5 = MNK(np.log(h_5[:4]), np.log(err_5[:4]))
print('Slope coefficient 5:', k_5)

plt.scatter(np.log(h_4[:-1]) / np.log(10), np.log(err_4[:-1]) / np.log(10), label = 'order 4')
plt.scatter(np.log(h_5[:-1]) / np.log(10), np.log(err_5[:-1]) / np.log(10), label = 'order 5')
plt.xlabel('lg(h)')
plt.ylabel('lg(err)')
plt.legend()
```
