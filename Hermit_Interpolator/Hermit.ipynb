```python
import numpy as np
import matplotlib.pyplot as plt

def aver(data):
  return np.sum(data) / np.size(data)
def MNK(x, y):
  k = (aver(x * y) - aver(x) * aver(y)) / (aver(x**2) - aver(x)**2)
  b = aver(y) - k * aver(x)
  return k, b
```

```python
l = np.array([2 / 2**i for i in range(6)])
Err_1 = np.array([0.0136782, 0.000354637, 6.15087e-06, 9.85495e-08, 1.54944e-09, 2.42475e-11])
Err_2 = np.array([0.000232419, 1.49735e-06, 6.48262e-09, 2.59562e-11, 1.02045e-13, 3.94216e-16])
Err_3 = np.array([2.16241e-06, 3.47139e-09, 3.75716e-12, 3.96905e-15, 7.11237e-17, 4.64039e-17 ])
ERR = [Err_1, Err_2, Err_3]

log_l_range = np.linspace(-3, 0.9, 10)
for i in range(3):
    plt.scatter(np.log(l), np.log(ERR[i]), label = 'N = {0}'.format(i + 3))
    k, b = MNK(np.log(l[:4]), np.log(ERR[i][:4]))
    plt.plot(log_l_range , k * log_l_range + b)
    print('Slope coefficient {0}:'.format(i), k)

plt.xlabel('log(l)')
plt.ylabel('log(Err)')
plt.title('     ')
plt.legend()
plt.show()
```
